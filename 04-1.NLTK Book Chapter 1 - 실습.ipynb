**<span style="color:crimson">[Your Turn]</span>** text5에서 "lol"의 출현 횟수는? 이 단어가 텍스트 전체에서 차지하는 비중은?

함수로 만들기

````python
def lexical_diversity(text):
    return len(set(text)) / len(text)

def percentage(count, total):
    return 100 * count / total
````

````python
lexical_diversity(text3)
````

````python
lexical_diversity(text5)
````

````python
percentage(text4.count('a'), len(text4))
````

# 3.2   Fine-grained Selection of Words

- 이번에는 긴 단어를 보자. 아마도 더 도움이 될지도 모른다.
- 길이가 15글자보다 긴 단어들의 집합을 다음과 같이 정의한다.  
    <span style="font-size:1.5em; line-height:36px">𝑃(𝑤)={𝑤|len(𝑤)>15}</span>
- 조사 대상 문서의 어휘를 _V_ 라고 하면, 길이 15 이상인 단어의 집합 _S_ 는  
    <span style="font-size:1.5em; line-height:36px">𝑆={𝑤│𝑤∈𝑉 𝑎𝑛𝑑 𝑤∈𝑃(𝑤)}</span>


````python
V = set(text1)
long_words = [w for w in V if len(w) > 15]  # [주의] 리스트이므로 중복될 수 있다
sorted(long_words)
````

# 3.3   Collocations and Bigrams

- **Collocation(연어)** : 유난히 함께 출현하는 단어열.  
  - collocation의 예: 'red wine', 'bright idea', 'talk freely'
  - collocation이 아닌 예 : 'the wine', 'this idea'   
- collocation의 특성: Native speaker에게 자연스럽게 보이는 조합이다. 따라서 어떤 한 단어를 비슷한 뜻의 단어로 치환하면 뜻이 매우 부자연스럽게 느껴진다.

|natural English...	|unnatural English... |
|-------------------|:-------------------:|
|the fast train     |the ~~quick~~ train|
|fast food          |~~quick~~ food|
|a quick shower     |a ~~fast~~ shower|
|a quick meal       |a ~~fast~~ meal|   

- Collocation은 **frequent bigram**이므로 collocation 을 처리하기 위해서는 우선 텍스트에서 단어 쌍(즉, **bigram**)들을 추출해야 한다.

````python
list(bigrams(['more', 'is', 'said', 'than', 'done']))
````

````python
bigrams(['more', 'is', 'said', 'than', 'done'])
````

### Functions Defined for NLTK's Frequency Distributions
|Example|Description|
|:-------|:-----------|
|fdist = FreqDist(samples)	|create a frequency distribution containing the given samples|
|fdist[sample] += 1	        |increment the count for this sample|
|fdist['monstrous']	        |count of the number of times a given sample occurred|
|fdist.freq('monstrous')	|frequency of a given sample|
|fdist.N()	                |total number of samples|
|fdist.most_common(n)	    |the n most common samples and their frequencies|
|for sample in fdist:	    |iterate over the samples|
|fdist.max()	            |sample with the greatest count|
|fdist.tabulate()	        |tabulate the frequency distribution|
|fdist.plot()	            |graphical plot of the frequency distribution|
|fdist.plot(cumulative=True)|cumulative plot of the frequency distribution|
|fdist1 |= fdist2	        |update fdist1 with counts from fdist2|
|fdist1 < fdist2	        |test if samples in fdist1 occur less frequently than in fdist2|
